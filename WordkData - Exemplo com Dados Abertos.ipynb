### ETAPA 1: Carregando as bibliotecas e os dados

```python
# Importando as bibliotecas que serão utilizadas no exemplo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

% matplotlib inline
```

```python
# URL - Deputados
dataset = "dados/deputados.csv"
```

### ETAPA 2: Analisando os dados

```python
# Obter dados
df_deputados = pd.read_csv(dataset, delimiter=";", low_memory=False)
# Possibilita a visualização de alguns registros do Dataset
df_deputados.head()
```

```python
# Avaliando os dados
df_deputados.columns
```

```python
# Análise descritiva dos dados
df_deputados.describe()
```

```python
# Existe algum dado faltante?
df_deputados.isnull().any()
```

```python
# Quantos dados faltantes existem por coluna?
dados_faltantes = df_deputados.isnull().sum()
dados_faltantes
```

```python
# Quantos dados faltantes em sua totalidade?
print ('Total de dados faltantes:', dados_faltantes.sum())
porcentagem = (dados_faltantes.sum()/ np.product(df_deputados.shape)) * 100
print ('Em um percentual de: %.2f' %  porcentagem)
```

### ETAPA 3: Reduzindo as dimensões

```python
# Selecionando as dimensões necessárias para o meu problema
df_deputados = df_deputados[['txNomeParlamentar', 'txtDescricao', 'vlrLiquido']]
```

```python
# Análise descritiva dos dados
df_deputados.describe()
```

```python
# Quem gastou mais
print(df_deputados[df_deputados.vlrLiquido == df_deputados['vlrLiquido'].max()]
)
print('*****************************************')
# Quem gastou menos
print(df_deputados[df_deputados.vlrLiquido == df_deputados['vlrLiquido'].min()])
```

```python
# Quem mais gastou com... :)
df_deputados[df_deputados.txtDescricao == 'DIVULGAÇÃO DA ATIVIDADE PARLAMENTAR.'].sort_values('vlrLiquido', ascending=False).head(10)
```

### ETAPA 4: Utilizando o conceito de Pivot Table para agrupar os dados

```python
# Criando uma Pivot Table
df_deputados.pivot_table(
        index=['txNomeParlamentar', 'txtDescricao'], 
        values=['vlrLiquido'], 
        aggfunc={'vlrLiquido': np.sum}
)
```

### Etapa 5: Selecione o deputado que você deseja analisar as suas despesas

```python
# Obtendo informações de um deputado qualquer (qualquer == você escolhe, ok?)
deputado_qualquer = df_deputados[ df_deputados['txNomeParlamentar'] == 'JAIR BOLSONARO'] 
# Visualiza informações sobre o deputado escolhido
deputado_qualquer.head()
```

### Etapa 6: Apresentandos os gastos em um gráfico

```python
# Atribuo o pivot table a uma variável
deputado_stack = deputado_qualquer.pivot_table(
        index=['txNomeParlamentar', 'txtDescricao'], 
        values=['vlrLiquido'], 
        aggfunc={'vlrLiquido': np.sum}
)


# Instancio o meu gráfico
fig, ax = plt.subplots(figsize=(10, 10))
ax.pie(deputado_stack, labels=["{1}".format(l2, l2) for l1, l2 in deputado_stack.index], autopct='%1.1f%%')
plt.title("Despesas do meu deputado")
```
